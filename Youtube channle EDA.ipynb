```python
import pandas as pd 
import numpy as mp
import seaborn as sns
import  matplotlib.pyplot  as plt
import matplotlib.ticker as ticker

%matplotlib inline
```

```python
df = pd.read_csv("Cleaned_data.csv")
```

```python
df.head()
```

```python
plt.figure(figsize = (12,6))
ax = sns.barplot( data = df.sort_values('viewCount', ascending=False)[0:9], x= 'title', y = 'viewCount')
plot = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos:'{:,.0f}'.format(x/1000) + 'K'))
# for i in ax.containers:
#     ax.bar_label(i,)
```

## Worst performing video

```python
ax = sns.barplot(x = 'title', y = 'viewCount', data = df.sort_values('viewCount', ascending=True)[0:9])
plot = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos:'{:,.0f}'.format(x/1000) + 'K'))
```

## Views, likes and comments

```python
fig, ax = plt.subplots(1,2)
sns.scatterplot(data = df, x = 'commentCount', y = 'viewCount', ax = ax[0])
sns.scatterplot(data = df, x = 'likeCount', y = 'viewCount', ax = ax[1])
```

## Duration of video

```python
sns.histplot(data = df, x = 'durationSecs', bins=30)
```

## Upload schedule 

```python
plt.figure(figsize= (12,6))
day_df = pd.DataFrame(df['pushblishDayName'].value_counts())
weekdays = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_df = day_df.reindex(weekdays)
ax = sns.barplot(data = day_df, x=day_df.index, y='pushblishDayName')
for i in ax.containers:
    ax.bar_label(i,)
```
